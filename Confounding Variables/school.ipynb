```python
import pandas as pd

# Load the data from the uploaded Excel file
file_path = 'public_data.xlsx'
data = pd.read_excel(file_path)
data.head()
```

```python
new_columns = [col.split()[1] if len(col.split()) > 1 else col for col in data.columns]

# Rename the columns
data.columns = new_columns
data
```

```python
melted_data = pd.melt(data, id_vars=['State'], var_name='Year', value_name='Enrollment')

# Create 'State_Year' identifier
melted_data['State_Year'] = melted_data['State'] + ', ' + melted_data['Year']

# Drop the original 'State' and 'Year' columns
melted_data = melted_data.drop(columns=['State', 'Year'])

# Pivot the table to have 'State_Year' as the index and 'Enrollment' as columns
pivot_data = melted_data.pivot_table(index='State_Year', values='Enrollment', aggfunc='sum').reset_index()

# Check the resulting DataFrame
pivot_data
```

```python
pivot_data.to_excel('public_enrollment_data.xlsx', index=False)
```

Private_enrollment

```python
import pandas as pd

# Load the data from the uploaded Excel file
file_path = 'private_data.xlsx'
data = pd.read_excel(file_path)
new_columns = [col.split()[1] if len(col.split()) > 1 else col for col in data.columns]

# Rename the columns
data.columns = new_columns
data['State'] = data['State'].apply(lambda x: x.split()[0])

melted_data = pd.melt(data, id_vars=['State'], var_name='Year', value_name='Enrollment')

# Create 'State_Year' identifier
melted_data['State_Year'] = melted_data['State'] + ', ' + melted_data['Year']

# Drop the original 'State' and 'Year' columns
melted_data = melted_data.drop(columns=['State', 'Year'])

# Pivot the table to have 'State_Year' as the index and 'Enrollment' as columns
pivot_data = melted_data.pivot_table(index='State_Year', values='Enrollment', aggfunc='sum').reset_index()

# Check the resulting DataFrame
pivot_data
```

```python
pivot_data.to_excel('private_enrollment_data.xlsx', index=False)
```
